```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory = './csvFiles_IK'
```

```python
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
```

```python
df_list = []
for file in csv_files:
    file_path = os.path.join(directory, file)
    df_list.append(pd.read_csv(file_path))
```

```python
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
combined_df
```

```python
len(combined_df)
```

```python
rotation_columns = [col for col in combined_df.columns if 'Rot' in col and 'Spine2' not in col]
rotation_df = combined_df[rotation_columns]
```

```python
rotation_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 스케일러 초기화
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
# 회전 데이터 피팅 및 변환
rotation_df_scaled = scaler.fit_transform(rotation_df)
```

```python
# 변환된 데이터를 DataFrame으로 변환
rotation_df_scaled = pd.DataFrame(rotation_df_scaled, columns=rotation_df.columns)
```

```python
train = rotation_df_scaled.iloc[:33641]
test = rotation_df_scaled.iloc[33641:]
```

```python
print(train.shape)
```

```python
len(train)
```

```python
len(test)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
train[:5]
```

```python
n_input = 10 
n_features = 63
batch_size = 5

generator = TimeseriesGenerator(rotation_df_scaled.values, rotation_df_scaled.values, length=n_input, batch_size=batch_size)
```

```python
len(train)
```

```python
len(generator)
```

```python
X, y = generator[0]
```

```python
y
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, BatchNormalization, Dense, TimeDistributed
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from keras.layers import Bidirectional
```

```python
model = Sequential()

# 첫 번째 양방향 LSTM 층
model.add(Bidirectional(LSTM(150, activation='relu', return_sequences=True), input_shape=(n_input, n_features)))
model.add(Dropout(0.2))

# 두 번째 LSTM 층
model.add(LSTM(100, activation='relu', return_sequences=False))
model.add(Dropout(0.2))

# 배치 정규화 층
model.add(BatchNormalization())

# 추가 Dense 층
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.2))

# 또 다른 Dense 층
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.2))

# 출력층
model.add(Dense(21 * 3))

# 모델 컴파일
model.compile(optimizer=Adam(learning_rate=0.001, decay=1e-6), loss='mse')

# 콜백 정의
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)
```

```python
model.summary()
```

```python
model.fit(generator, epochs=50, callbacks=[early_stopping, reduce_lr, model_checkpoint])
```

```python
model.save('final_model_IK.h5')
```

```python
myloss = model.history.history['loss']
plt.plot(range(len(myloss)),myloss)
```

```python
# 예측 값을 넣을 빈 리스트
test_predictions = []
# 훈련 데이터셋에서 마지막 입력 개수의 값을 가져온 후
first_eval_batch = train[-n_input:]
first_eval_batch = first_eval_batch.to_numpy()
# LSTM이 원하는 형식으로 데이터를 reshape (TimeSeriesGenerator와 같은 형식)
current_batch = first_eval_batch.reshape((1, n_input, n_features))

# 어디까지 예측할 것인지 결정하는 부분 (지금은 12개의 데이터를 예측하므로 test 데이터셋의 길이만큼)
for i in range(len(test)):
    # 마지막 포인트보다 한 시점 뒤를 예측
    current_pred = model.predict(current_batch)[0] # 예측값의 []를 한쌍 없애기 위해 [0]

    # 예측값을 predictions 리스트에 추가
    test_predictions.append(current_pred)

    # 현재 배지가 예측값을 포함하도록 '업데이트'
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```
